<h1>3.2 Calculating team performance</h1>

<h3>By calculating the overall team performance we can analze the data easier and the classifier can make decisions better based on the performance of each team</h3>

<h3>My approach to calculating the performance is by averaging the history for every team and using the number of goals that were succesful or not for both attacking and defending </h3>

```python
dataset.columns
```

```python
# Team - each team from EPL
# HG - goals scored at home
# AG - goals scored away
# HAtk - home attack performance
# AAtk - away attack performance
# HDef - home defense performance
# ADef - away defense performance
# HGFail - goals failed to defend at home
# AGFail - goals failed to defend away
team_performance = pd.DataFrame(columns=('Team','HG','AG','HAtk','AAtk', 'HDef','ADef','HGFail','AGFail'))
team_home = dataset.groupby('HomeTeam')
team_away = dataset.groupby('AwayTeam')
```

```python
#calculate the number of goals scored at home and away 
home_scored = dataset.FTHG.sum() * 1.0 / dataset.shape[0]
away_scored = dataset.FTAG.sum() * 1.0 / dataset.shape[0]
#the failed goals are based on the assumption that
#scored goals away for a team are failed goals at home for the opponents
#and vice versa
home_failed = away_scored
away_failed = home_scored
```

```python
#calculate the number of total games
total_games = dataset.shape[0]
total_games
```

```python
#list all teams from the match data
team_performance.Team = team_home.HomeTeam.all().values

#calculate the home goals and away goals
team_performance.HG = team_home.FTHG.sum().values
team_performance.AG = team_away.FTAG.sum().values

#calculate the home and away failed-to-defend goals
team_performance.HGFail = team_home.FTAG.sum().values
team_performance.AGFail = team_away.FTHG.sum().values

#calculate the home and away attack performance
#this is calculated based on the total number of games and the number of goals succeeded
team_performance.HAtk = (team_performance.HG / total_games) / home_scored
team_performance.AAtk = (team_performance.AG / total_games) / away_scored

#calculate the home and away defense performance
#this is calculated based on the total number of games and the number of failed-to-defend goals
team_performance.HDef = (team_performance.HGFail / total_games) / home_failed
team_performance.ADef = (team_performance.AGFail / total_games) / away_failed

team_performance
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
teams = team_performance['Team']
y_pos = np.arange(len(teams))
performance = team_performance['HAtk']
error = np.random.rand(len(teams))

ax.barh(y_pos, performance, xerr=error, align='edge',
        color='blue', ecolor='black')
ax.set_yticks(y_pos)
ax.set_yticklabels(teams)
ax.invert_yaxis() 
ax.set_xlabel('Attack Performance')
ax.set_title('Team Attack Performance')

plt.show()
```

<h1>3.3 Generate Feature Table and fill the 2018 Dataset</h1>

<h3>This will be the table that contains all the features needed for training the model</h3>

<h3>The features that I will use are:</h3>
<ul>
<li>HST and AST - on a lot of sources, especially Kaggle, people were saying that HST and AST are really good for predicting the outcomes</li>
<li>HAtk and AAtk - performance of attack for both teams</li>
<li>HDef and ADef - performance of defense for both teams</li>
</ul>
<p>Having the performances as features will make it easier for determining whether a team is better than the other</p>

```python
features = dataset[['HomeTeam','AwayTeam','FTR','HST','AST']]

#initialise the arrays for adding the performance for each team
HAtk = []
AAtk = []
HDef = []
ADef = []

HAtk2018 = []
HST2018 = []
AST2018 = []
AAtk2018 = []
HDef2018 = []
ADef2018 = []
```

```python
dataset_2018 = dataset18[['HomeTeam','AwayTeam','FTR']]
```

```python
#iterate through the features table and add performances if the teams are equal in the team_performance table
for index,row in features.iterrows():    
    HAtk.append(team_performance[team_performance['Team'] == row['HomeTeam']]['HAtk'].values[0])
    AAtk.append(team_performance[team_performance['Team'] == row['AwayTeam']]['AAtk'].values[0])
    HDef.append(team_performance[team_performance['Team'] == row['HomeTeam']]['HDef'].values[0])
    ADef.append(team_performance[team_performance['Team'] == row['AwayTeam']]['ADef'].values[0])
```

```python
#do the same for the 2018 data
for index,row in dataset_2018.iterrows():
    sumHome = 0
    sumAway = 0
    counter = 0
    
    #loop through the feature table to sum up the values of HST and AST
    for j,row2 in features.iterrows():
        if row['HomeTeam'] == row2['HomeTeam'] and row['AwayTeam'] == row2['AwayTeam']:
            sumHome = sumHome + row2['HST']
            sumAway = sumAway + row2['AST']
            counter = counter + 1
     
    #I calculate HST and AST as averages here for every match between 2 specific teams
    #for example, if Arsenal and Tottenham have a match in 2018, then the HST and AST
    #are averages of every HST and AST in the past Arsenal-Tottenham matches
    #this gives a general idea of how most of the matches between these 2 teams played out
    HST2018.append(round(sumHome / counter))
    AST2018.append(round(sumAway / counter))
    HAtk2018.append(team_performance[team_performance['Team'] == row['HomeTeam']]['HAtk'].values[0])
    AAtk2018.append(team_performance[team_performance['Team'] == row['AwayTeam']]['AAtk'].values[0])
    HDef2018.append(team_performance[team_performance['Team'] == row['HomeTeam']]['HDef'].values[0])
    ADef2018.append(team_performance[team_performance['Team'] == row['AwayTeam']]['ADef'].values[0])
```

```python
#update the features table and the 2018 dataset
features['HAtk'] = HAtk
features['AAtk'] = AAtk
features['HDef'] = HDef
features['ADef'] = ADef

dataset_2018['HST'] = HST2018
dataset_2018['AST'] = AST2018
dataset_2018['HAtk'] = HAtk2018
dataset_2018['AAtk'] = AAtk2018
dataset_2018['HDef'] = HDef2018
dataset_2018['ADef'] = ADef2018
```

<h1>4. Methodology Overview</h1>

<p><font size="4">I started by reading research papers on the matter and looking at Kaggle march madness to get ideas of how other people have gone and solved problems like this before. I found some good hints on which I based my idea, especially in the IJMTER and Yandex papers. From those I got the idea to use team performance as a feature, though some used more complicated features such as motivation, composition and form. 
After documenting enough on predicting match results, I have decided to use the following methodology:</font></p>
<p><font size="5">1) Add new data</font></p>
<p><font size="4">I added the new season data so that I would get a more accurate team performance based on recent results.</font></p>
<p><font size="4"></font></p>
<br>
<font size="5">2) Create the feature table</font>
<p><font size="4">After experimenting on the available data and reading some papers, I chose my features that would be used for training the model. I started by applying a Pearson correlation on the datasets and saw which fields influence the FTR more. Moreover, a lot of people on Kaggle were saying that adding HST and AST greatly improves the model, so I decided to have these 2 as my first features.<br>
Then I decided to engineer my own features since statistics like red and yellow cards or fouls don't really influence the result that much.<br>
I chose to assess the performance of each team for attacking and defending. I looked at FIFA data first, but only found player statistics, which I didn't want to add as it could've confused the model. So, I engineered those features by myself.<br>
My main assumption for calculating the performance is by looking at the number of goals scored by each team both home and away.
For the attack performance, I calculate how many goals a team managed to give on average across the whole data. For the defense performance, I look at how many goals the team failed to defend.
This data should give the model a feel for how good a team is and help it make better predictions.<br>
Besides that, I also added all the features to the 2018 set. For HST and AST, I looked at the history between two teams and averaged the values for each of them. I applied that for every combination of 2 teams.
</font></p>
<br>
<p><font size="5">3) Encode categorical data</font></p>
<p><font size="4">Since I am predicting strings (H, D and A), I had to encode the data. I had a look at OneHotEncoder, but ended up using LabelEncoder. OneHotEncoder would have created 3 different columns, one for each result, and if a result was a Draw for example, then the columns would have been populated with 1 on the Draw column and 0 on the others. I went back to LabelEncoder due to the fact that FTR is the field I am predicting, and I am not sure if prediciting 3 columns would have worked well. Therefore, with LabelEncoder, I encode HomeWin as 0, Draws as 1 and AwayWins as 2.</font></p>

<br>
<p><font size="5">4) Apply classification methods on the dataset</font></p>
<p><font size="4">Since I didn't know exactly what model would work well, I tried multiple examples. Someone on Kaggle said XGBoost was a really good model for predicting this data so I installed and tried it out as well. These are the models I applied:<br>
Random Forest, Multinomial Naive Bayes, XGBoost, Logistic Regression, Decision Tree Classification, K-nearest Neightbours, Linear SVM and Gradient Boosting.<br>
I splitted the data into a train set and a test set to perform cross validation, and chose 70% for my train set and 30% for testing purposes. After fitting all of the classifiers and predicting the data, I obtained the accuracy for every method. I looked at their percentage and whether they predicted all outcomes or not (since some algorithms with good percentages didn't predict draws).<br>
Since the best looking one that predicted all outcomes was XGBoost, I decided to take it further and tune its parameters.
</font></p>

<br>
<p><font size="5">5) Apply a neural net</font></p>
<p><font size="4">Before getting to the phase of tuning XGBoost, I wanted to try and apply a neural network to see whether I can obtain even better results. I first tried the Multilayer Perceptron network built in scikit and obtained a good accuracy, but the biggest issue was that it wouldn't predict any draws. So, I tried defining my own neural network in Keras, using Tensorflow as backend. I made a basic classification neural network and trained it on the train data. However, this approach was unsuccesful, as I only got around 30% performance. Most likely, I would need to look more into defining the layers and what activation functions to use.</font></p>

<br>
<p><font size="5">6) Tuning the parameters</font></p>
<p><font size="4">Before tuning the parameters, I performed a k-fold cross validation to see how well the model performed on different test sets. After that, I implemented GridSearch and tried various parameter configurations for XGBoost. After finding the best parameters, I tried fitting the classifier again and obtained a small improvement in the accuracy. I wouldn't say the improvement was significant, but probably experimenting with multiple configurations would give even better results. For now, I only tried with popular values for every parameter.</font></p>

<br>
<p><font size="5">7) Predicting 2018 results</font></p>
<p><font size="4">Finally, after tunning the parameters and getting my final classifier, I predicted the results on the 2018 dataset. While I didn't get any draw outcomes, I believe the results are plausible since it shows that the best team (at least in my view and based on the performances) wins in every match.</font></p>

<h1>5.1 Applying the classification on the data</h1>

<img src="images/classification.png">

```python
#trying multiple classifier methods
clf1 = RandomForestClassifier()
clf2 = MultinomialNB()
clf3 = XGBClassifier()
clf4 = LogisticRegression()
clf5 = DecisionTreeClassifier()
clf6 = KNeighborsClassifier()
clf7 = LinearSVC()
clf8 = GradientBoostingClassifier()
```

```python
#assess the performance of every classifier
x_axis = ["RNF", "MNB", "XGB", "LR", "DT", "KNN", "LSVM", "GB"]
y_axis =  [accuracy_score(y_test, y_pred1), accuracy_score(y_test, y_pred2), accuracy_score(y_test, y_pred3),
                            accuracy_score(y_test, y_pred4), accuracy_score(y_test, y_pred5), accuracy_score(y_test, y_pred6),
                            accuracy_score(y_test, y_pred7), accuracy_score(y_test, y_pred8)]
```

```python
plt.bar(x_axis, y_axis)
plt.show()
pd.DataFrame({'Model' :x_axis, 'Score' : y_axis})
```

<h3>As it can be clearly seen, XGB has the best performance out of all algorithms and can also predict all the outcomes, so we will use it later and try to tune the parameters to obtain better results</h3>
